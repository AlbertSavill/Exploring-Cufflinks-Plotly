```python
df = pd.DataFrame(np.random.randn(100,3), columns = ['A', 'B', 'C'])
df.head()
df['A'] = df['A'].cumsum() + 20
df['B'] = df['B'].cumsum() + 20
df['C'] = df['C'].cumsum() + 20
```

```python
df.shape
```

```python
df.iplot()
```

```python
plt.plot(df)
```

```python
df.plot()
```

```python
df.iplot(x = 'A', y = 'B', mode = 'markers', size = 12)
```

```python
titanic = sns.load_dataset('Titanic')
titanic.head()
```

```python
titanic.iplot(kind = 'bar', x = 'sex', y = 'survived', title = 'Survived', xTitle = 'Sex', yTitle = 'Survived')
```

```python
df3 = pd.DataFrame({'X': [10,20,30,20,10],
                    'Y': [10,20,30,20,10],
                    'Z': [10,20,30,20,10]})
df3.head()
```

```python
df3.iplot(kind = 'surface', colorscale = 'rdylbu')
```

```python
df.iplot(kind = 'surface', colorscale = 'rdylbu')
```

```python
cf.datagen.sinwave(10, 0.25).iplot(kind = 'surface')
```

```python
cf.datagen.scatter3d(2, 150, mode = 'stocks').iplot(kind = 'scatter3d', x = 'x', y = 'y', z = 'z')
```

```python
df[['A', 'B']].iplot(kind = 'spread')
```

```python
df.iplot(kind = 'hist', bins = 25, barmode = 'group', bargap = 0.5)
```

```python
cf.datagen.bubble3d(5,4,mode='stocks').iplot(kind='bubble3d', x = 'x', y = 'y', z = 'z', size = 'size')
```

```python
cf.datagen.heatmap(20,20).iplot(kind = 'heatmap', colorscale = 'spectral', title = 'Cufflinks - Heatmap')
```
